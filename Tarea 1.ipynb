```python
#Importación de librerías
#import torch as torch
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
# if using a Jupyter notebook, include:
%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/lsalazarv/Tarea1/master/daily-min-temperatures.csv'
df = pd.read_csv(url,index_col=0,skip_blank_lines=False)
#obtenemos la temperatura minima y maxima
tempMin = df['Temp'].min()
tempMax = df['Temp'].max()
#obtenemos la desviacion estandar del set de datos
vdesvstandar = df['Temp'].std()
#obtenemos la media del set de datos
vmedia = df['Temp'].mean()
#se declara una lista para guardar los datos normalizados y otra para guardar los datos estandarizados
normalizacion_list = []
estandarizacion_list=[]
# se recorre el set de datos y se realiza la normalizacion de los datos
for i in range(len(df)):
    resp1 = df.iloc[i]['Temp']-tempMin
    resp2 = tempMax-tempMin
    resp3 = resp1/resp2
    normalizacion_list.append(resp3)
    
    vrespnorm1 = df.iloc[i]['Temp'] - vmedia
    vrespnorm2 = vrespnorm1/vdesvstandar
    estandarizacion_list.append(vrespnorm2)
    #print(i,df.iloc[i]['Temp'])

#print(vdesvstandar,vmedia) 
print('datos normalizados')
print(normalizacion_list) 
print('datos estandarizados')
print(estandarizacion_list)
#print(normalizacion_list)   

```
